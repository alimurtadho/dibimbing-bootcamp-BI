```python
# Preprocessing
```

```python
df.isnull().sum()
```

```python
# mengcopy data

df_ulik = df.copy()

# df_ulik = df ?
```

```python
# drop kolom
df_ulik = df.drop(columns='Cabin')
```

```python
df_ulik.head()
```

```python
# fill na untuk mengisi missing values

df_ulik['Age_fill_by_20'] = df_ulik['Age'].fillna(20)
```

```python
df_ulik.isnull().sum()
```

```python
rerata_usia = df_ulik['Age'].mean()

# min, max, count
```

```python
df_ulik['Age_fill_by_mean'] = df_ulik['Age'].fillna(rerata_usia)
```

```python
# replace dengan notasi dictionary

df_ulik['Pclass_cat'] = df_ulik['Pclass'].replace(
    {
        1:'Eksekutif',
        2:'Bisnis',
        3:'Ekonomi'
    }
)
```

```python
df_ulik.head()
```

```python
# pandas binning

df_ulik['Age_cat'] = pd.cut(
    x=df_ulik['Age_fill_by_mean'],
    bins=[0,5,11,16,25,35,45,55,65,100],
    labels=['balita','kanak-kanak','remaja 1','remaja 2','dewasa 1','dewasa 2','lansia 1','lansia 2','manula']
)
```

```python
df_ulik.head()
```

```python
# mengganti tipe data

# membulatkan nilai
df_ulik['Age_round'] = df_ulik['Age_fill_by_mean'].round()
```

```python
df_ulik.head()
```

```python
df_ulik['Age_int'] = df_ulik['Age_round'].astype('int')
```

```python
df_ulik.head()
```

```python
# filtering
# df[kondisi]

df_ulik[df_ulik['Embarked']=='S']
```

```python
df_ulik[(df_ulik['Embarked']=='S') & (df_ulik['Survived'] == 1)]
```
