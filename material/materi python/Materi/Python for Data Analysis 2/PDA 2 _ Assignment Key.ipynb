```python
import pandas as pd
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vT1enHtD-_jxEIuctKQaAhpK2b-Jl66hqskpFni7tUOcFDR6YLNSQrRznKQczSyrsGq_cSJFMN4wn-u/pub?gid=0&single=true&output=csv')
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df.head()
```

# 1. Nilai penjualan tahun 2017

```python
df[df['order_date'].dt.year == 2017]['sales'].sum()
```

# 2. Kenaikan penjualan dari 2015 ke 2016

```python
sales_by_year = pd.pivot_table(
    data=df,
    index=df['order_date'].dt.year,
    values='sales',
    aggfunc='sum'
)

sales_by_year.pct_change() * 100
```
